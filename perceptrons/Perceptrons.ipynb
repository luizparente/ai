# Perceptrons

This example tests the implementation of a perceptron as provided in the `Perceptron` class. We start by selecting a linearly separable toy dataset, preparing it for training, running the perceptron, and assessing the results. 

## Dataset Selection

Since the perceptron is a binary classifier, the dataset must feature two classes that can be separated by a linear boundary. As we can see below, the Iris dataset is linearly separable, and can thus be used in this demo.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Loading the Iris dataset.
iris = load_iris()
X = iris.data
# Pre-processing data as Setosa and Non-Setosa classes (binary classes).
y = (iris.target != 0).astype(int)

# Selecting features: sepal length (index 0) and sepal width (index 1).
sepal_length = X[:, 0]
sepal_width = X[:, 1]

# Plotting the data to check linear separability.
plt.figure(figsize=(8, 6))
plt.scatter(sepal_length[y == 0], sepal_width[y == 0], color='blue', marker='o', label='Setosa (class 0)')
plt.scatter(sepal_length[y == 1], sepal_width[y == 1], color='red', marker='x', label='Non-Setosa (class 1)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset: Setosa vs Non-Setosa - Linear Separability')
plt.legend()
plt.grid(True)
plt.show()
```

## Perceptron Predictions

After confirming the iris dataset is suitable for binary classification and linear separability, we can move on to the next step, and use the `Perceptron` class to classify the data.

```python
from Perceptron import Perceptron
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Loading Iris dataset.
iris = load_iris()
X = iris.data  # Features.
y = (iris.target != 0).astype(int)  # Binary labels: 0 for Setosa, 1 for others.

# Splitting into train and test sets (75% train, 25% test, stratified).
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

# Initializing and training the perceptron on train set (4 features).
perceptron = Perceptron(num_inputs=4, learning_rate=0.1, epochs=50)
perceptron.train(X_train, y_train)

# Testing predictions on the test set.
predictions = [perceptron.predict(inputs) for inputs in X_test]

# Calculating accuracy.
accuracy = np.mean(predictions == y_test) * 100

print(f"Accuracy on test set: {accuracy:.2f}%")
```
